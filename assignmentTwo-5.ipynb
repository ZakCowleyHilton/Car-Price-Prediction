# **ANOVA Analysis of Categorical Predictors**

```python
import statsmodels.formula.api as sm
import statsmodels.api as sm
from statsmodels.formula.api import ols

# List of categorical predictors that i want to analyse in an ANOVA style
categorical_predictors = ['Manufacturer', 'Category', 'Color', 'Model', 'Year']

# writing a Function to perform ANOVA analysis and print results
def perform_anova_for_predictor(predictor):
    # Correctly reference the predictor column within the formula string
    model = ols(f'Price ~ {predictor}', data=df).fit()
    anova_table = sm.stats.anova_lm(model, typ=2)
    print(f"ANOVA results for Price vs {predictor}:")
    print(anova_table)
    print("\n")

# Perform ANOVA for each categorical predictor
for predictor in categorical_predictors:
    perform_anova_for_predictor(predictor)
```

# Random Forest Regressor:

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import MinMaxScaler
from sklearn import metrics
import matplotlib.pyplot as plt
from sklearn.metrics import make_scorer

# Create a new DataFrame as a copy of the original to help with version control
df_ml = df.copy()

# duplicate row deleting
df_ml = df_ml.drop_duplicates()

# Select predictors and target variable
SelectedColumns = ['Category', 'Year', 'Manufacturer']
TargetVariable = 'Price'

# Prepare data for ML
DataForML = df_ml[SelectedColumns]
DataForML_Numeric = pd.get_dummies(DataForML, drop_first=True)
DataForML_Numeric[TargetVariable] = df_ml[TargetVariable]

# Separate Predictors and Target Variable
X = DataForML_Numeric.drop(TargetVariable, axis=1).values
y = DataForML_Numeric[TargetVariable].values

# Scaling the predictors
PredictorScaler = MinMaxScaler()
X_scaled = PredictorScaler.fit_transform(X)

# Split the data into training and testing set
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)

# Random Forest Regressor
RegModel = RandomForestRegressor(max_depth=4, n_estimators=400, criterion='friedman_mse')

# Printing all the parameters of Random Forest
print(RegModel)

# Creating the model on Training Data
RF = RegModel.fit(X_train, y_train)

# Predicting on the test data
predictions = RF.predict(X_test)

# Measuring fit in training data
print('R2 Value on training data:', metrics.r2_score(y_train, RF.predict(X_train)))

# Plotting the feature importance for Top 10 most important columns
%matplotlib inline
feature_importances = pd.Series(RF.feature_importances_, index=DataForML_Numeric.drop(TargetVariable, axis=1).columns)
feature_importances.nlargest(10).plot(kind='barh')
plt.show()

print('\n##### Model Validation and Accuracy Calculations ##########')

# Printing some sample values of prediction
TestingDataResults = pd.DataFrame(data=X_test, columns=DataForML_Numeric.drop(TargetVariable, axis=1).columns)
TestingDataResults[TargetVariable] = y_test
TestingDataResults['Predicted' + TargetVariable] = np.round(predictions, 2)

# Printing sample prediction values
print(TestingDataResults.head())

# Calculating the error for each row
TestingDataResults['APE'] = 100 * (np.abs(TestingDataResults[TargetVariable] - TestingDataResults['Predicted' + TargetVariable]) / TestingDataResults[TargetVariable])

# Calculating Mean and Median Absolute Percentage Error
MAPE = np.mean(TestingDataResults['APE'])
MedianMAPE = np.median(TestingDataResults['APE'])

# Calculating Accuracies
Accuracy = 100 - MAPE
MedianAccuracy = 100 - MedianMAPE
print('Mean Accuracy on test data:', Accuracy)
print('Median Accuracy on test data:', MedianAccuracy)

# Defining a custom function to calculate accuracy
def Accuracy_Score(orig, pred):
    MAPE = np.mean(100 * (np.abs(orig - pred) / orig))
    # print('#'*70,'Accuracy:', 100-MAPE)
    return 100 - MAPE

# Custom Scoring MAPE calculation
custom_Scoring = make_scorer(Accuracy_Score, greater_is_better=True)

# Running 10-Fold Cross validation on a given algorithm
Accuracy_Values = cross_val_score(RegModel, X, y, cv=10, scoring=custom_Scoring)
print('\nAccuracy values for 10-fold Cross Validation:\n', Accuracy_Values)
print('\nFinal Average Accuracy of the model:', round(Accuracy_Values.mean(), 2))
```

```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import MinMaxScaler
from sklearn import metrics
import matplotlib.pyplot as plt
from sklearn.metrics import make_scorer
from IPython.display import Image
from sklearn import tree
import pydotplus

# Create a new DataFrame as a copy of the original and remove duplicates
df_ml = df.copy().drop_duplicates()

# Select predictors and target variable
SelectedColumns = ['Category', 'Year', 'Manufacturer']
TargetVariable = 'Price'

# Prepare data for machine learning
DataForML = df_ml[SelectedColumns]
DataForML_Numeric = pd.get_dummies(DataForML, drop_first=True)
DataForML_Numeric[TargetVariable] = df_ml[TargetVariable]

# Separate Predictors and Target Variable
X = DataForML_Numeric.drop(TargetVariable, axis=1).values
y = DataForML_Numeric[TargetVariable].values

# Scale the predictors
PredictorScaler = MinMaxScaler()
X_scaled = PredictorScaler.fit_transform(X)

# Split the data into training and testing set
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)

# Decision Tree Regressor
RegModel = DecisionTreeRegressor(max_depth=5, criterion='friedman_mse')

# Printing all the Decision Tree
print(RegModel)

# Creating the model on Training Data
DT = RegModel.fit(X_train, y_train)

# Predicting on the test data
predictions = DT.predict(X_test)

# Measuring the fit in Training data
print('R2 Value on training data:', metrics.r2_score(y_train, DT.predict(X_train)))

# Plotting the feature importance for Top 10 most important columns
%matplotlib inline
feature_importances = pd.Series(DT.feature_importances_, index=DataForML_Numeric.drop(TargetVariable, axis=1).columns)
feature_importances.nlargest(10).plot(kind='barh')
plt.show()

print('\n##### Model Validation and Accuracy Calculations ##########')

# Printing some sample values of prediction
TestingDataResults = pd.DataFrame(data=X_test, columns=DataForML_Numeric.drop(TargetVariable, axis=1).columns)
TestingDataResults[TargetVariable] = y_test
TestingDataResults['Predicted' + TargetVariable] = np.round(predictions, 2)

# Printing sample prediction values
print(TestingDataResults.head())

# Calculating the error for each row
TestingDataResults['APE'] = 100 * (np.abs(TestingDataResults[TargetVariable] - TestingDataResults['Predicted' + TargetVariable]) / TestingDataResults[TargetVariable])

# Calculating Mean and Median Absolute Percentage Error
MAPE = np.mean(TestingDataResults['APE'])
MedianMAPE = np.median(TestingDataResults['APE'])

# Calculating Accuracies
Accuracy = 100 - MAPE
MedianAccuracy = 100 - MedianMAPE
print('Mean Accuracy on test data:', Accuracy)
print('Median Accuracy on test data:', MedianAccuracy)

# Defining a custom function to calculate accuracy
def Accuracy_Score(orig, pred):
    MAPE = np.mean(100 * (np.abs(orig - pred) / orig))
    return 100 - MAPE

# Custom Scoring MAPE calculation
custom_Scoring = make_scorer(Accuracy_Score, greater_is_better=True)

# Running cross validation on a given algorithm
Accuracy_Values = cross_val_score(RegModel, X, y, cv=10, scoring=custom_Scoring)
print('\nAccuracy values for 10-fold Cross Validation:\n', Accuracy_Values)
print('\nFinal Average Accuracy of the model:', round(Accuracy_Values.mean(), 2))

# Visualizing the Decision Tree
dot_data = tree.export_graphviz(RegModel, out_file=None,
                                feature_names=DataForML_Numeric.drop(TargetVariable, axis=1).columns)
graph = pydotplus.graph_from_dot_data(dot_data)
Image(graph.create_png(), width=2000, height=2000)
```

```python
import pandas as pd
import pickle

def FunctionPredictResultLogistic(InputData):
    # Load the training data structure for column alignment
    DataForML = pd.read_pickle('DataForML.pkl')

    # One-hot encode the new input data (the same way as your training data)
    InputData = pd.get_dummies(InputData, drop_first=True)

    # Determine the column names without the target column for prediction
    if 'Binary_Target' in DataForML.columns:
        predictors = DataForML.drop('Binary_Target', axis=1).columns
    else:
        predictors = DataForML.columns

    # Align input columns with the columns used during the training
    InputData = InputData.reindex(columns=predictors, fill_value=0)

    # Load the logistic regression model from the pickle file
    with open('logistic_model_full_with_manufacturer.pkl', 'rb') as fileReadStream:
        PredictionModel = pickle.load(fileReadStream)

    # Generate predictions
    Prediction = PredictionModel.predict(InputData)
    PredictionResult = pd.DataFrame(Prediction, columns=['Prediction'])

    return PredictionResult

# Creating the function which takes inputs and returns prediction
def FunctionGenerateCarPrediction(inp_Category, inp_Year, inp_Manufacturer):
    # Create a DataFrame for the model input
    SampleInputData = pd.DataFrame(
        data=[[inp_Category, inp_Year, inp_Manufacturer]],
        columns=['Category', 'Year', 'Manufacturer']
    )

    # Call the prediction function with aligned input data
    Predictions = FunctionPredictResultLogistic(InputData=SampleInputData)

    # Return predictions in JSON format
    return Predictions.to_json()

# Example function call with inputs
result_json = FunctionGenerateCarPrediction(
    inp_Category='Sedan',
    inp_Year=2020,
    inp_Manufacturer='Ford'
)

print("Prediction Result (in JSON format):")
print(result_json)
```

```python
import pandas as pd
import tkinter as tk
from tkinter import messagebox, ttk
from sklearn.linear_model import LogisticRegression
import pickle

# Data pre-processing: Load and preprocess the dataset before using it in your application.
df = pd.read_csv('car_price_prediction.csv')
df.rename(columns={'Prod. year': 'Year'}, inplace=True)
columns_to_remove = ['Odo', 'ID', 'Mileage', 'Model', 'Color', 'Leather interior',
                     'Levy', 'Drive wheels', 'Doors', 'Wheel', 'Airbags', 'Gear box type',
                     'Cylinders', 'Fuel type', 'Engine volume']
df.drop(columns=columns_to_remove, errors='ignore', inplace=True)

print(df.head())
print(df.columns)


class CarPricePredictionApp:
    def __init__(self, master, data):
        self.master = master
        self.master.title('Car Price Prediction')

        # Use preprocessed data
        self.data = data

        # Predictor columns and prepare data with one-hot encoding
        self.predictor_columns = ['Category', 'Year', 'Manufacturer']
        self.X = pd.get_dummies(self.data[self.predictor_columns], drop_first=True)

        # Load the pre-trained model
        with open('logistic_model_full_with_manufacturer.pkl', 'rb') as f:
            self.model = pickle.load(f)

        self.create_widgets()

    def create_widgets(self):
        # Creating a dropdown for Category
        category_label = tk.Label(self.master, text='Category:')
        category_label.grid(row=0, column=0)
        self.category_combobox = ttk.Combobox(self.master, values=self.data['Category'].unique())
        self.category_combobox.grid(row=0, column=1)

        # Creating entry for Year
        year_label = tk.Label(self.master, text='Year:')
        year_label.grid(row=1, column=0)
        self.year_entry = tk.Entry(self.master)
        self.year_entry.grid(row=1, column=1)

        # Creating a dropdown for Manufacturer
        manufacturer_label = tk.Label(self.master, text='Manufacturer:')
        manufacturer_label.grid(row=2, column=0)
        self.manufacturer_combobox = ttk.Combobox(self.master, values=self.data['Manufacturer'].unique())
        self.manufacturer_combobox.grid(row=2, column=1)

        # Button to predict price
        predict_button = tk.Button(self.master, text='Predict Price', command=self.predict_price)
        predict_button.grid(row=3, columnspan=2)

    def predict_price(self):
        # Capture input data
        category = self.category_combobox.get()
        year = self.year_entry.get()
        manufacturer = self.manufacturer_combobox.get()

        # Prepare the input data
        input_data = pd.DataFrame([[category, year, manufacturer]], columns=self.predictor_columns)
        input_data = pd.get_dummies(input_data)

        # Ensure the input has all the columns the model was trained with
        for col in self.X.columns:
            if col not in input_data.columns:
                input_data[col] = 0
        input_data = input_data.reindex(columns=self.X.columns, fill_value=0)

        # Print out the transformed input data for debugging
        print("Transformed Input Data:")
        print(input_data)

        # Make prediction
        price_prediction = self.model.predict(input_data.values)[0]
        probability = self.model.predict_proba(input_data.values)[0][1]

        # Display results
        result_text = 'Expensive' if price_prediction else 'Affordable'
        messagebox.showinfo('Prediction Details',
                            f'Prediction: {result_text}\n'
                            f'Probability of being Expensive: {probability:.2f}')

if __name__ == '__main__':
    root = tk.Tk()
    app = CarPricePredictionApp(root, df)
    root.mainloop()
```
